**Credit Card Fraud Detection**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_df = pd.read_csv('creditcard.csv')
credit_df.head()
```

```python
credit_df['Class'].value_counts()
```

```python
credit_df.describe()
```

```python
legit = credit_df[credit_df.Class == 0]
fraud = credit_df[credit_df.Class == 1]
```

```python
fraud.shape
```

```python
legit_sample = legit.sample(n=len(fraud), random_state=2)
credit_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
credit_df.groupby('Class').mean()
```

```python
x = credit_df.drop(columns='Class', axis=1)
y = credit_df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
train_acc = accuracy_score(model.predict(x_train), y_train)
test_acc = accuracy_score(predictions, y_test)
```

```python
print('Training Accuracy: ', train_acc)
print('Testing Accuracy: ', test_acc)
```
